```python
def generate_data(max_step, n_0, temps_mu, N_max, p, ftns):
    fitness_dict = {}
    
    g_wt = hg.Genotype(ftns, hg.our_first_sequence) 


    index_experiment = []

    for t,mu in enumerate(temps_mu):
        
        pop_mu = {}
        list_mutants = [copy.copy(g_wt) for i in range(n_0)]

        for step in range(max_step):

            list_mutants_aux = []
            for x in list_mutants:
                node = copy.copy(x)
                n = int(node.fitness) #number of descendants of each node

                for i in range(n):
                    #n_muts = int(np.random.poisson(node.L*mu))
                    n_muts = 1

                    if n_muts == 0:
                        list_mutants_aux.append(node)
                        index_experiment.append(node.sequence)
                    else:
                        node_m = node.mutate(n_muts=n_muts)
                        try:
                            node_m.fitness = fitness_dict[node_m.sequence]
                            if node_m.fitness != 0:
                                list_mutants_aux.append(node_m)
                                
                            else:
                                list_mutants_aux.append(node)
                                
                                
                        except:
                            r = random.random()
                            if r<=p:
                                node_m.fitness = ftns
                                fitness_dict[node_m.sequence] = ftns
                                list_mutants_aux.append(node_m)
                                index_experiment.append(node_m.sequence)
                            else:
                                node_m.fitness = 0
                                fitness_dict[node_m.sequence] = 0
                                list_mutants_aux.append(node)
                                index_experiment.append(node.sequence)                              
                                            
                    #End of mutants in population loop
            
            
            
            #Truncate the population
            if ftns > 1:
                if len(list_mutants_aux) < N_max:
                    pass
                else:
                    random.shuffle(list_mutants_aux)
                    list_mutants_aux = list_mutants_aux[:N_max]
            else:
                if len(list_mutants_aux) != N_max:
                    print(f'Something wrong: len mutants list:{len(list_mutants_aux)} and N_max={N_max}')

            
            list_mutants = copy.copy(list_mutants_aux)   #For the next step, we use list_mutants   
                
            step +=1
            if step<10:
                step = '0' + str(step)
            else:
                step = str(step)
            if t==0:
                name = f'30-1-{step}-EL3_all_all_trim_merged_filter_sort_filter_length_collapsed.fasta'
            elif t==1:
                name = f'c43-p{step}-3-EL3_all_all_trim_merged_filter_sort_filter_length_collapsed.fasta'
            hg.save_pop(list_mutants, f'./Data1/Sequences_filtered/{name}')

            
    print('mean fitness', np.array(list(fitness_dict.values())).mean(), 'size=', len(list_mutants))
    #Create index for the created data

    index_experiment = list(set(index_experiment))

    with open('./Data1/seqs_index.dict', 'w') as fw:
        for i, seq in enumerate(index_experiment):

            fw.write(f'{i}\t{seq}\n')

    with open('./Data1/seqs_fitness.dict', 'w') as fw:

        for i, seq in enumerate(fitness_dict.keys()):

            fw.write(f"{seq}\t{fitness_dict[seq]}\n")
            
            
def analyze_fname(a):
	'''
	Function that looks in file_name the values of
	temperature and step
	
	Returns:  
		- temperature (Celsius degrees)
		- step
	'''
	if a[0] == 'c':
		step = int(a.split('-')[1][1:])
		t=43
	elif a[0] == '3':
		step = int(a.split('-')[2])
		t=30
	return t, step

def calculateParametersLSR(x, y):
	"""
	Function that calculates w* and b* for a least square regression and the standard error for b*
	Inputs:
		x: array containing the x-values in log-scale
		y: array containing the y-values in log-scale
	Outputs:
		b: float containing the intercept term of the regression line
		w: float containing the slope of the regression line
		SEw: float containing the standard error of b*
	"""

	meanTarget = np.mean(y)
	meanFeature = np.mean(x)

	centeredTarget = y - meanTarget
	centeredFeature = x - meanFeature

	w = (centeredFeature @ centeredTarget)/(centeredFeature @ centeredFeature)

	b = meanTarget - w *  meanFeature

	# Standard error
	yHat = b + w * x
	n = x.shape[0]

	SEw = np.sqrt((1/(n - 2)) * ((np.sum(np.power((y - yHat), 2)))/(np.sum(np.power((x - meanFeature), 2)))))



	return b, w, SEw

greek_leeterz = [chr(code) for code in range(945,970)]
#Select the wt
wt_sequence='CAACAAGGTCAGCTATATCATAATATCGATATTGTAGACGGCTTTGACAGACGTGACATCCGGCTCAAATCTTTCACCATAAAAGGTGAACGAAATGGGCGGCCTGTTAACGTTTCTGCTAGCCTGTCTGCTGTCGATTTATTTTACAGCCGACTCCATACGAGCAATCTTCCGTTCGCTACACTAGATCTTGATACTACCTTTAGTTCGTTTAAACACGTTCTTGATAGTATCTTTTTATTAACCCAACGCGTAAAGCGTTGAAACTTTG'
temps = [30,43]
cmap = {30: 'blue', 43:'red'}
symbdict = {2:'o',12:'s', 30:'^', 58:'*'}
nLoad = 1000

def distance(seq, seq_ref):
    '''
        Function that returns the Humming distance between
        two sequences
        Input:
            -sequence to compare
            -reference sequence
        Return:
            -int : distance
    '''
    d=0
    for char in range(len(seq_ref)):
        if seq[char] != seq_ref[char]:
            d+=1
    return d

def compute_alpha(dir_path):
    
    results = {t: [] for t in temps}


    index_name =  dir_path + 'seqs_index.dict'

    index_dict = {}
    inv_index_dict = {}

    with open(index_name, 'r') as f:
        for line in f:
            L = line.split('\t')
            index_dict[L[1][:-1]] = int(L[0])
            inv_index_dict[int(L[0])] = L[1][:-1]



    datapath = dir_path + 'Sequences_filtered/'
    #Read all fileNames in folder
    allFileNames = [f for f in listdir(datapath) if isfile(join(datapath, f))]
    allFileNames.sort()

    d_results_list = {t:{} for t in temps}

    for thisFileName in allFileNames:
        t,step = analyze_fname(thisFileName)
        d_results_list[t][step] = []

        with open(os.path.join(datapath, thisFileName), 'r') as f:
            lines = f.readlines()
            for i,line in enumerate(lines):
                if line[0] == '>':
                    id = int(line.split('-')[0].replace('>', ''))
                    abundance=int(line.split('-')[1].replace('\n', ''))
                    sequence = lines[i+1].replace('\n', '')
                    #print(id)
                    d = distance(sequence,hg.our_first_sequence)
                    for n in range(abundance):
                        d_results_list[t][step].append(d)


    #Active this to plot d^2 dependence with time
    if 1==1:
        for t in temps:
            L = list(d_results_list[t].keys())
            L.sort()
            y = []
            y_err = []
            y_sq = []
            index = []
            step_list=[]

            #Compute moments of d
            for step in L:
                if step > 15:
                    step_list.append(step)
                    y_sq.append((np.array(d_results_list[t][step])**2).mean())

            #Do fitting for steps >25
            b,w,E = calculateParametersLSR(np.log2(np.array(step_list)), np.log2(np.array(y_sq)))
            results[t] = [w,E]
        
    return results
```

```python
#Number of sequences at each step for mu = 0
max_step = 50
max_rep = 1

n_0 = 10

temps_mu = [10**-4,10**-3]
N_max = 10**4

p = 1.0
f = 1

dir_path = './Data1/'

results = {}


for rep in range(max_rep):
    results[rep] = {}
    ps = []
    for i,p in enumerate(np.logspace(-3,0, 3, base=10.0)):
    #if 1==1:
        p=1.0
        print(rep+1, 'p=',p)
        ps.append(p)
        #Generate data
        if f == 1:
            print('Constant size population')
            n_0 = N_max
            
        generate_data(max_step, n_0, temps_mu, N_max, p, f)

        #Compute alpha
        results[rep][p] = compute_alpha(dir_path)
```

```python
#How does the abundance evolves?
#Number of sequences at each step for mu = 0
max_step = 30
max_rep = 1

n_0 = 3

mu = 10**-5
N_max = 10**4




pop_mu = {}
abund_distr = {i:{} for i in range(max_rep)}

population_size = {i:[] for i in range(max_rep)}
for repetition in range(max_rep):

    list_mutants = [g_wt for i in range(n_0)]
    population_size[repetition].append(len(list_mutants))

    for step in range(max_step):
        list_mutants_aux = []
        for node in list_mutants:
            n = int(node.fitness)#number of descendants of each node
            
            for i in range(n):
                n_muts = int(np.random.poisson(node.L*mu))
            
                if n_muts !=0:
                    node = node.mutate(n_muts=n_muts,exclude_lethal=False)   
                list_mutants_aux.append(node)
        

        list_mutants = list_mutants_aux


        #Save the population size for control
        population_size[repetition].append(len(list_mutants))
        
        #Searching abundances and normalize               
        abundance = {}

        for genotype in list_mutants:
            try:
                abundance[genotype.sequence] +=1
            except:
                abundance[genotype.sequence] = 1

        abund_values = list(abundance.values())
        abund_values.sort(reverse=True)

        normalization = 0
        for element in abund_values:
            normalization += element

        abund_values = [a/normalization for a in abund_values]
        
        abund_distr[repetition][step] = abund_values



        #Truncate the population
        if len(list_mutants)<N_max:
            pass
        else:
            random.shuffle(list_mutants)
            list_mutants = list_mutants[:N_max]
        
```

```python
for rep in list(abund_distr.keys())[:1]:
    for step in abund_distr[rep].keys():
        if step in [1,5, 10, 15, 20, 25, 29]:
            x = np.arange(1,len(abund_distr[rep][step])+1)
            y = np.array([abund_distr[rep][step][i] for i in range(len(x))])
            plt.plot(x,y, 'o-', label=f'step {step}')
        
plt.yscale('log')
plt.xscale('log')
plt.legend()
plt.ylabel('relative abundance')
plt.xlabel('rank')
plt.title(f'mu={mu}')
plt.savefig(f'./abundance_evolution_mu_{mu}_withlethal.jpg', format='jpg')
```

```python
r = 1.0
mu = 0.01
N_max = 10**4
```

```python

#This script plots the abundances for a cycle in different moments
max_t = 15
max_it = 1

plt.figure()

for m in range(max_it):
    list_mutants = [g_wt for i in range(3)]
    for t in range(max_t):
        print(f'Replication cycle of the system:{t+1} of {max_t}, starting number of sequences={len(list_mutants)}')

        for node in list_mutants:
            n = np.random.poisson(node.fitness*r)
            #print(n)
            for i in range(n):
                if random.random() <= mu:
                    list_mutants.append(node.mutate(exclude_lethal=True))
                else:
                    list_mutants.append(node)
        if len(list_mutants)<N_max:
            pass
        else:
            random.shuffle(list_mutants)
            list_mutants = list_mutants[:N_max]
        
        print(f'Analizing the system with size={len(list_mutants)}')
    
    
    
        #Searching abundances and normalize               
        abundance = {}

        for genotype in list_mutants:
            try:
                abundance[genotype.sequence] +=1
            except:
                abundance[genotype.sequence] = 1

        abund_values = list(abundance.values())

        abund_values.sort(reverse=True)

        normalization = 0
        for element in abund_values:
            normalization += element

        abund_values = [a/normalization for a in abund_values]
        plt.plot([1+x for x in range(len(abund_values))], abund_values, 'o-', label=f'generation {t+1} size={len(list_mutants)}')
        
        #print('Saving population')
    
        #save_pop(list_mutants, f'./Data_simulated/{m}_{t}')

    
    
    
    

plt.yscale('log')
plt.xscale('log')
plt.legend(loc='best')



plt.xlabel('rank')
plt.ylabel('relative abundance')
plt.savefig('./abundance_progressive_1.jpg', format='jpg')
```

```python
list_sequences = [g.sequence for g in list_mutants]
```

```python
list_sequences = list(set(list_sequences))
print(len(list_sequences))

def distance(seq, seq_ref):
    '''
        Function that returns the Humming distance between
        two sequences
        Input:
            -sequence to compare
            -reference sequence
        Return:
            -int : distance
    '''
    d=0
    for char in range(len(seq_ref)):
        if seq[char] != seq_ref[char]:
            d+=1
    return d
```

```python
adj = {}

for i, a in enumerate(list_sequences):
    print(i)
    for j,b in enumerate(list_sequences[i:]):
        if distance(b,a) ==1:
            #print(f'First neighbor of {a}: {b}')
            try: adj[i].append(j)
            except: adj[i] = []
```

```python
import networkx as nx

G = nx.Graph()

for k in adj.keys():
    if len(adj[k]):
        for n in adj[k]:
            G.add_edge(int(k),n)
    else:
        G.add_node(int(k))
```

```python
nx.draw(G, node_size=4)
plt.savefig('genotype.jpg')
```

```python
G0 = G.subgraph(max(nx.connected_components(G), key=len))
```

```python
nx.draw(G0, node_size=4)
plt.savefig('gc.jpg')
```

```python
k_list = [G0.degree(n) for n in G0.nodes()]
k_dict = {}


for k in k_list:
    try: k_dict[k] +=1/len(k_list)
    except: k_dict[k] =1/len(k_list)
```
