```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
L=271

#m = [3*L//4+10]
m =[0]



t_max = 10*L


def f(value, L):
    return value + (1-value/(3*L))*(1-(4*value)/(3*L))
    

for t in range(1,t_max):
    m.append(f(m[t-1], L))
```

```python

x = np.arange(0,t_max)
y = np.array(m)

plt.plot(x, y, 'o-')
plt.plot(x,[3/4*L for i in x], '--')
plt.plot(x[:L],x[:L], '--')

plt.yscale('log')
plt.xscale('log')
```

```python
plt.plot(y,[f(i, L)-i for i in m], 'o-')
```
