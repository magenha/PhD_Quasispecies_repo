```python
import numpy as np
import matplotlib.pyplot as plt

# Function definition for m(x)
def m(x, lam):
    return lam * (1 - np.exp(-x / lam))

# Parameters
lam = 7000
p = 0.9

# Function to calculate the expression
def calculate_expression(t):
    first_term = p**t * m(t, lam)
    second_term = sum(p**(t-i) * (1-p) * m(t-i, lam) for i in range(0, t + 1))
    return first_term + second_term

# Values for t from 2 to 10
t_values = np.arange(0, 100)

# Calculate the expression for each value of t
expression_values = [calculate_expression(t) for t in t_values]

# Plotting
plt.plot(t_values, expression_values, marker='o')
plt.plot(t_values, t_values)
plt.xlabel('t')
plt.ylabel('Expression')
plt.title(f'p^t * m(t) + Sum(i=1 to t, p^(t-i)*(1-p)*m(t-i))\n(lambda={lam}, p={p})')
plt.grid(True)

plt.yscale('log')
plt.xscale('log')
```
