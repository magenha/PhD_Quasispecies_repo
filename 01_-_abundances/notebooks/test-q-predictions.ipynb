```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

r_max = 10**5

x = np.linspace(1,r_max,r_max)
```

```python
beta = 1.5
k_min = 2
q = 1.3
y = np.power(1.0-(1-q)*beta*(x-k_min),1/(1-q))/np.power(1.0-(1-q)*beta*(x-k_min),1/(1-q)).sum()

plt.plot(x,y)
plt.plot(x,np.power(x,-1.0*beta)/np.power(x,-1.0*beta).sum())
plt.xscale('log')
plt.yscale('log')

plt.ylim(1.0*10**-7, 1)
```
